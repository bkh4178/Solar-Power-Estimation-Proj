```python
# ==============================================
# Step 1 ‚Äî Import & Feature Definition
# ==============================================
import pandas as pd
import numpy as np
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt
import seaborn as sns
import joblib

# ‚úÖ ÏµúÏ¢Ö ÏÇ¨Ïö©Ìï† Feature Î¶¨Ïä§Ìä∏
final_feats = [
    'humidity', 'wind_gust_spd', 'hour', 'doy', 'wind_spd_b',
    'ceiling', 'uv_idx', 'appr_temp', 'uv_cloud_adj', 'dow',
    'hour_sin', 'doy_sin', 'is_rain', 'rain', 'hour_cos',
    'doy_cos', 'snow', 'coord1', 'coord2', 'haze'
]
target_col = "nins"
print("‚úÖ Final feature count:", len(final_feats))
```

```python
# ==============================================
# Step 2 ‚Äî Define add_features() function
# ==============================================
def add_features(df):
    EPS = 1e-6
    df["time"] = pd.to_datetime(df["time"])

    # 1Ô∏è‚É£ ÏãúÍ∞Ñ Í¥ÄÎ†® ÌååÏÉù
    df["hour"] = df["time"].dt.hour
    df["dow"] = df["time"].dt.dayofweek
    df["doy"] = df["time"].dt.dayofyear
    df["hour_sin"] = np.sin(2 * np.pi * df["hour"] / 24)
    df["hour_cos"] = np.cos(2 * np.pi * df["hour"] / 24)
    df["doy_sin"] = np.sin(2 * np.pi * df["doy"] / 365)
    df["doy_cos"] = np.cos(2 * np.pi * df["doy"] / 365)

    # 2Ô∏è‚É£ Íµ¨Î¶Ñ Í¥ÄÎ†®
    if "cloud_a" in df.columns:
        df["cloud_a_norm"] = df["cloud_a"] / (df["cloud_a"].max() + EPS)
        df["uv_cloud_adj"] = df["uv_idx"] * (1 - df["cloud_a_norm"])
    else:
        df["uv_cloud_adj"] = df["uv_idx"]

    # 3Ô∏è‚É£ Í∞ïÏàò Ïó¨Î∂Ä
    if "precip_1h" in df.columns:
        df["is_rain"] = (df["precip_1h"] > 0).astype("int8")
    else:
        df["is_rain"] = 0

    # 4Ô∏è‚É£ Í∞ÄÏãúÍ±∞Î¶¨ ‚Üí haze
    if "vis" in df.columns:
        df["haze"] = 1 / (df["vis"] + EPS)
    else:
        df["haze"] = 0

    return df
```

```python
# ==============================================
# Step 3 ‚Äî Load CSV & Apply add_features (Î≥¥Í∞Ñ Ï∂îÍ∞Ä)
# ==============================================
print("üöÄ Loading train.csv ...")
df = pd.read_csv("train.csv", low_memory=True, memory_map=True)
print(f"‚úÖ Loaded shape: {df.shape}")

df = add_features(df)

available_feats = [f for f in final_feats if f in df.columns]
missing_feats = [f for f in final_feats if f not in df.columns]
print(f"‚öôÔ∏è Available features: {len(available_feats)} / Missing: {missing_feats}")

# ‚úÖ --- [ÏàòÏ†ïÎêú interpolate_groupwise Ìï®Ïàò] ---
def interpolate_groupwise(df):
    df = df.sort_values(["pv_id", "time"])
    num_cols = df.select_dtypes(include=["float32", "float64", "int32", "int64"]).columns
    for col in num_cols:
        df[col] = df.groupby("pv_id")[col].transform(
            lambda x: x.interpolate(method="linear", limit_direction="both")
        )
        df[col] = df.groupby("pv_id")[col].transform(lambda x: x.bfill().ffill())
    return df

# ‚úÖ Î≥¥Í∞Ñ Ï†ÅÏö©
if "pv_id" in df.columns and "time" in df.columns:
    df = interpolate_groupwise(df)
else:
    print("‚ö†Ô∏è Warning: 'pv_id' or 'time' missing ‚Äî skip interpolation")

# ‚úÖ ÌïÑÏöîÌïú Ïª¨ÎüºÎßå ÎÇ®Í∏∞Í∏∞
df = df[available_feats + [target_col]].astype("float32")

print(f"‚úÖ Final shape after interpolation: {df.shape}")
print(f"‚úÖ Null count after interpolation: {df.isna().sum().sum()}")
```

```python
# ==============================================
# Step 4 ‚Äî Split Data
# ==============================================
X = df.drop(columns=[target_col])
y = df[target_col].values

sample_frac = 0.07  # Ï†ÑÏ≤¥ Ï§ë ÏïΩ 7%
X_sample, _, y_sample, _ = train_test_split(X, y, train_size=sample_frac, random_state=42)
print(f"‚öôÔ∏è Sampled {len(X_sample):,} rows (‚âà {sample_frac*100:.1f}%)")

X_train, X_valid, y_train, y_valid = train_test_split(
    X_sample, y_sample, test_size=0.2, random_state=42
)
print(f"‚úÖ Train: {len(X_train):,} | Valid: {len(X_valid):,}")
```

```python
# ==============================================
# Step 5 ‚Äî Train XGBoost
# ==============================================
model = XGBRegressor(
    n_estimators=1000,
    learning_rate=0.04,
    max_depth=7,
    subsample=0.8,
    colsample_bytree=0.8,
    min_child_weight=4,
    random_state=42,
    tree_method="hist",
    n_jobs=-1
)

print("\nüöÄ Training XGBoost...")
model.fit(X_train, y_train)
print("‚úÖ Model training complete")
```

```python
# ==============================================
# Step 6 ‚Äî Evaluate Model
# ==============================================
preds = model.predict(X_valid)
mae = mean_absolute_error(y_valid, preds)
print(f"‚úÖ Validation MAE: {mae:.4f}")
```

```python
# ==============================================
# Step 7 ‚Äî Save Model & Importance
# ==============================================
feat_imp = pd.DataFrame({
    "feature": X_train.columns,
    "importance": model.feature_importances_
}).sort_values("importance", ascending=False)

print("\nüèÜ Top 10 Important Features:")
display(feat_imp.head(10))

plt.figure(figsize=(9, 7))
sns.barplot(data=feat_imp.head(10), x="importance", y="feature", palette="viridis")
plt.title("Feature Importance (Top 10)")
plt.tight_layout()
plt.show()

feat_imp.to_csv("feature_importance_final20.csv", index=False, encoding="utf-8-sig")
joblib.dump(model, "xgb_final20.pkl")
print("üíæ Saved ‚Üí xgb_final20.pkl, feature_importance_final20.csv")
```

```python
# ==============================================
# Step 9-1 ‚Äî Full-data train + test Î°úÎìú + Î≥¥Í∞Ñ + ÌÉúÏñëÍ∏∞Ìïò
# ==============================================
print("üöÄ Loading train & test CSVs ...")

import numpy as np
import pandas as pd
from astral import LocationInfo
from astral.sun import sun
import datetime as dt

# --- ‚ë† train/test Î°úÎìú ---
train = pd.read_csv("train.csv", low_memory=True, memory_map=True)
test  = pd.read_csv("test.csv", low_memory=True, memory_map=True)
print(f"‚úÖ Train shape: {train.shape} | Test shape: {test.shape}")

# --- ‚ë° Í∏∞Î≥∏ ÌååÏÉùÎ≥ÄÏàò Ï∂îÍ∞Ä ---
train = add_features(train)
test  = add_features(test)

# --- ‚ë¢ Î∞úÏ†ÑÏÜåÎ≥Ñ Î≥¥Í∞Ñ(interpolation) ---
def interpolate_groupwise(df):
    df = df.sort_values(["pv_id", "time"])
    num_cols = df.select_dtypes(include=["float32", "float64", "int32", "int64"]).columns
    for col in num_cols:
        df[col] = df.groupby("pv_id")[col].transform(
            lambda x: x.interpolate(method="linear", limit_direction="both")
        )
        df[col] = df.groupby("pv_id")[col].transform(lambda x: x.bfill().ffill())
    return df

print("üîÑ Interpolating train/test ...")
train = interpolate_groupwise(train)
test  = interpolate_groupwise(test)

# --- ‚ë£ ÌÉúÏñë Í≥†ÎèÑ/Î∞©ÏúÑÍ∞Å/ÏΩîÏÇ¨Ïù∏Í≥†ÎèÑ Ï∂îÍ∞Ä ---
print("üåû Adding solar geometry features...")

def add_solar_geometry(df):
    # ÌèâÍ∑† ÏúÑÎèÑ¬∑Í≤ΩÎèÑ (Î∞úÏ†ÑÏÜå Ï†ÑÏ≤¥Ïùò ÌèâÍ∑† Ï¢åÌëú ÏÇ¨Ïö©)
    lat, lon = df["coord1"].mean(), df["coord2"].mean()
    loc = LocationInfo(latitude=lat, longitude=lon)

    # üåû Í∞Å ÏãúÍ∞ÅÎ≥Ñ ÌÉúÏñë Í≥†ÎèÑ¬∑Î∞©ÏúÑÍ∞Å Í≥ÑÏÇ∞
    def get_solar_values(t):
        if pd.isnull(t):
            return (np.nan, np.nan)
        try:
            s = sun(observer=loc.observer, date=t)
            elev = s["noon"].elevation if hasattr(s["noon"], "elevation") else np.nan
            azim = s["noon"].azimuth if hasattr(s["noon"], "azimuth") else np.nan
            return (elev, azim)
        except Exception:
            return (np.nan, np.nan)

    df["elevation"], df["azimuth"] = zip(*df["time"].apply(get_solar_values))
    df["cosZ"] = np.cos(np.deg2rad(90 - df["elevation"]))
    return df

train = add_solar_geometry(train)
test  = add_solar_geometry(test)

# --- ‚ë§ feature list ÏóÖÎç∞Ïù¥Ìä∏ ---
final_feats += ["elevation", "azimuth", "cosZ"]

# --- ‚ë• ÌïÑÏöîÌïú Ïª¨ÎüºÎßå ÎÇ®Í∏∞Í∏∞ ---
usecols = final_feats + [target_col]
train = train[usecols].fillna(0).astype("float32")
test  = test[[c for c in final_feats if c in test.columns]].fillna(0).astype("float32")

print(f"‚úÖ After interpolation & solar features ‚Äî Train: {train.shape}, Test: {test.shape}")


```

```python
# ==============================================
# Step 9-2 ‚Äî Full-data ÌïôÏäµ
# ==============================================
X_full = train.drop(columns=[target_col])
y_full = train[target_col].values

final_model = XGBRegressor(
    n_estimators=2500,       # Ï°∞Í∏à ÎäòÎ¶º (Full data ÏïàÏ†ïÌôîÏö©)
    learning_rate=0.04,
    max_depth=7,
    subsample=0.8,
    colsample_bytree=0.8,
    min_child_weight=4,
    random_state=42,
    tree_method="hist",
    n_jobs=-1,
    verbosity=1
)

print("\nüöÄ Full-data training start ...")
final_model.fit(X_full, y_full)
print("‚úÖ Full-data training complete")

# Ï†ÄÏû•
import joblib
joblib.dump(final_model, "xgb_full_final.pkl")
print("üíæ Saved ‚Üí xgb_full_final.pkl")
```

```python
# ==============================================
# Step 9-3 ‚Äî Test ÏòàÏ∏° + ÌõÑÏ≤òÎ¶¨
# ==============================================
print("üöÄ Predicting on test set...")
pred_test = final_model.predict(test)

# ÏùåÏàòÍ∞í ÌÅ¥Î¶¨Ìïë
pred_test = np.clip(pred_test, 0, None)

print(f"‚úÖ Prediction done. Range: {pred_test.min():.3f} ~ {pred_test.max():.3f}")
```

```python
# ==============================================
# Step 9-4 ‚Äî Save Submission
# ==============================================
sub = pd.read_csv("submission_sample.csv")
print(f"‚úÖ Loaded submission template: {sub.shape}")

sub["nins"] = pred_test.astype("float32")

# Ï§ëÎ≥µ Î∞è Í≤∞Ï∏° Í≤ÄÏÇ¨
print("Null count:", sub["nins"].isna().sum())
print("Duplicate keys:", sub.duplicated(subset=["time", "pv_id", "type"]).sum())

# Ï†ÄÏû•
sub.to_csv("result_submission.csv", index=False, encoding="utf-8-sig")
print("üíæ Saved ‚Üí result_submission.csv")
```
